# Dataset Guide: Column Descriptions / دليل أعمدة البيانات

**1. STUDENT IDENTIFIER**: Unique ID for each student; used to track individual records.  
**معرف الطالب**: رقم فريد لكل طالب لتتبع سجلاته الفردية.  

**2. STDNT_AGE**: Age of the student at enrollment.  
**عمر الطالب**: عمر الطالب عند التسجيل.  

**3. STDNT_GENDER**: Gender of the student (M/F).  
**جنس الطالب**: ذكر أو أنثى.  

**4. STDNT_BACKGROUND**: Coded category for student background or demographic group.  
**الخلفية أو المجموعة الديموغرافية**: تصنيف مشفر لخلفية الطالب أو مجموعته الديموغرافية.  

**5. IN_STATE_FLAG**: Indicates if the student is in-state (Y/N).  
**علم الطالب المقيم في الولاية**: يوضح إذا كان الطالب من نفس الولاية (نعم/لا).  

**6. INTERNATIONAL_STS**: Indicates if the student is an international student (Y/N).  
**حالة الطالب الدولي**: يوضح إذا كان الطالب دوليًا (نعم/لا).  

**7. STDNT_TEST_ENTRANCE_COMB**: Combined score from entrance tests for the student.  
**درجة اختبارات الدخول المجمعة**: مجموع درجات الطالب في اختبارات القبول.  

**8. FIRST_TERM**: code representing the first term of enrollment (e.g., Fall 2009).  
**الفصل الدراسي الأول**: رمز يمثل الفصل الدراسي الأول للتسجيل.  

**9. SECOND_TERM**: code representing the second term of enrollment (e.g., Spring 2010 ).  
**الفصل الدراسي الثاني**: رمز يمثل الفصل الدراسي الثاني للتسجيل.  

**10. HOUSING_STS**: Housing status of the student (On Campus/Off Campus).  
**حالة السكن**: سكن الطالب داخل الحرم الجامعي أو خارجه.  

**11. RETURNED_2ND_YR**: Indicates if the student returned for the second year (1 = Yes, 0 = No).  
**العودة للسنة الثانية**: يوضح إذا عاد الطالب للسنة الثانية (1=نعم، 0=لا).  

**12. DISTANCE_FROM_HOME**: Distance from student's home to campus (in miles or km).  
**المسافة من المنزل**: المسافة بين منزل الطالب والحرم الجامعي.  

**13. HIGH_SCHL_GPA**: Student’s high school GPA.  
**المعدل التراكمي للثانوية**: معدل الطالب في المدرسة الثانوية.  

**14. HIGH_SCHL_NAME**: Name or code of the high school attended.  
**اسم المدرسة الثانوية**: اسم المدرسة الثانوية التي التحق بها الطالب أو رمزها.  

**15. FATHER_HI_EDU_DESC**: Father’s highest education level description.  
**أعلى مستوى تعليمي للأب**: وصف أعلى مستوى تعليمي للأب.  

**16. MOTHER_HI_EDU_DESC**: Mother’s highest education level description.  
**أعلى مستوى تعليمي للأم**: وصف أعلى مستوى تعليمي للأم.  

**17. DEGREE_GROUP_DESC**: Type of degree program (e.g., Bachelors, Associate).  
**نوع البرنامج الدراسي**: نوع الدرجة الجامعية أو البرنامج التعليمي.  

**18. FIRST_TERM_ATTEMPT_HRS**: Total credit hours attempted in the first term.  
**عدد الساعات المعتمدة المقررة في الفصل الأول**: مجموع الساعات الدراسية التي حاول الطالب اجتيازها.  

**19. FIRST_TERM_EARNED_HRS**: Total credit hours successfully earned in the first term.  
**عدد الساعات المعتمدة المكتسبة في الفصل الأول**: مجموع الساعات التي اجتازها الطالب بنجاح.  

**20. SECOND_TERM_ATTEMPT_HRS**: Total credit hours attempted in the second term.  
**عدد الساعات المعتمدة المقررة في الفصل الثاني**.  

**21. SECOND_TERM_EARNED_HRS**: Total credit hours successfully earned in the second term.  
**عدد الساعات المعتمدة المكتسبة في الفصل الثاني**.  

**22. GROSS_FIN_NEED**: Total calculated financial need of the student.  
**إجمالي الحاجة المالية**: مجموع الحاجة المالية للطالب المحسوبة.  

**23. COST_OF_ATTEND**: Total cost of attendance (tuition, fees, housing, etc.).  
**تكلفة الحضور الكلية**: مجموع التكاليف الدراسية والسكنية وغيرها.  

**24. EST_FAM_CONTRIBUTION**: Estimated financial contribution from the family.  
**المساهمة المالية المقدرة من الأسرة**.  

**25. UNMET_NEED**: Portion of financial need not met by family or aid.  
**الحاجة المالية غير الملباة**: الجزء من الحاجة المالية الذي لم يتم تغطيته.  

**26. Fall_GPA**: GPA calculated for the first term.  
**المعدل التراكمي للفصل الأول**.  

**27. Spring_GPA**: GPA calculated for the second term.  
**المعدل التراكمي للفصل الثاني**.  

**28. Passed_Count**: Number of courses passed by the student in a term.  
**عدد المواد التي تم اجتيازها**.  

**29. Failed_Count**: Number of courses failed in a term.  
**عدد المواد التي رسب فيها الطالب**.  

**30. Incomplete_Count**: Number of courses marked incomplete in a term.  
**عدد المواد غير المكتملة**.  

**31. First_Completion_Rate**: Completion rate for the first term (`FIRST_TERM_EARNED_HRS / FIRST_TERM_ATTEMPT_HRS`).  
**معدل الإنجاز للفصل الأول**: نسبة الساعات المكتسبة إلى الساعات المقررة في الفصل الأول.  

**32. Second_Completion_Rate**: Completion rate for the second term (`SECOND_TERM_EARNED_HRS / SECOND_TERM_ATTEMPT_HRS`).  
**معدل الإنجاز للفصل الثاني**.  

**33. Total_Attempt_Hours**: Total credit hours attempted across both terms.  
**إجمالي الساعات المقررة للفصلين**.  

**34. Total_Earned_Hours**: Total credit hours earned across both terms.  
**إجمالي الساعات المكتسبة للفصلين**.  

**35. Total_Completion_Rate**: Overall completion rate (`Total_Earned_Hours / Total_Attempt_Hours`).  
**معدل الإنجاز الإجمالي**.  

**36. Need_Coverage_Ratio**: Ratio of estimated family contribution to cost of attendance (`EST_FAM_CONTRIBUTION / COST_OF_ATTEND`).  
**نسبة تغطية الحاجة**: نسبة مساهمة الأسرة المقدرة إلى تكلفة الحضور.  

**37. Unmet_Need_Ratio**: Ratio of unmet financial need to cost of attendance (`UNMET_NEED / COST_OF_ATTEND`).  
**نسبة الحاجة غير الملباة**.  

**38. Net_Cost**: Cost of attendance minus estimated family contribution (`COST_OF_ATTEND - EST_FAM_CONTRIBUTION`).  
**التكلفة الصافية**: تكلفة الحضور مطروحًا منها مساهمة الأسرة.  

**39. Financial_Difficulty**: Combined measure of financial burden (`GROSS_FIN_NEED + UNMET_NEED`).  
**الصعوبة المالية**: مقياس إجمالي للعبء المالي على الطالب.  

**40. Major_Category**: Grouped major category for analysis (e.g., STEM, Business, Arts & Humanities).  
**فئة التخصص الدراسي**: تصنيف التخصصات لتسهيل التحليل (مثل: علوم، إدارة أعمال، فنون وإنسانيات).
