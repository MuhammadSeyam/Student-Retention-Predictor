# What do we need to drop?

* Columns with high missing values 
* Columns that are useless for ML   (ID, High_School_Name)
* Duplicated columns                (FATHER_HI_EDU_DESC / MOTHER_HI_EDU_DESC -> MOTHER_HI_EDU_CD / FATHER_HI_EDU_CD	)

```python
# Removing High Null Columns 
d = df_na[df_na['NaN %'] > 50]['Column'].to_list()
df = df.drop(columns=d)
```

```python
# Removing useless columns
df = df.drop(columns= ['STUDENT IDENTIFIER','HIGH_SCHL_NAME'])
```

```python
# Removing Redundant Columns
df = df.drop(columns=['MOTHER_HI_EDU_DESC','FATHER_HI_EDU_DESC','DEGREE_GROUP_DESC','STDNT_TEST_ENTRANCE2'])
```

# 1. Filling numerical columns with the LOCAL MEDIAN

```python
df['STDNT_TEST_ENTRANCE_COMB'] = df.groupby('STDNT_MAJOR')['STDNT_TEST_ENTRANCE_COMB']\
                                .transform(lambda x:x.fillna(x.median()))
```

```python
df['DISTANCE_FROM_HOME'] = df.groupby('IN_STATE_FLAG')['DISTANCE_FROM_HOME']\
                                .transform(lambda x: x.fillna(x.median()))
```

```python
df['HIGH_SCHL_GPA'] = df.groupby('STDNT_BACKGROUND')['HIGH_SCHL_GPA']\
                                .transform(lambda x: x.fillna(x.median()))
```

```python
df['FATHER_HI_EDU_CD'] = df.groupby('STDNT_BACKGROUND')['FATHER_HI_EDU_CD']\
                                .transform(lambda x: x.fillna(x.median()))    
```

```python
df['MOTHER_HI_EDU_CD'] = df.groupby('STDNT_BACKGROUND')['MOTHER_HI_EDU_CD']\
                                .transform(lambda x: x.fillna(x.median()))    
```

```python
df['SECOND_TERM_ATTEMPT_HRS'] = df.groupby('DEGREE_GROUP_CD')['SECOND_TERM_ATTEMPT_HRS']\
                                .transform(lambda x: x.fillna(x.median()))    
```

```python
df['SECOND_TERM_EARNED_HRS'] = df.groupby('DEGREE_GROUP_CD')['SECOND_TERM_EARNED_HRS']\
                                .transform(lambda x: x.fillna(x.median()))    
```

```python
numeric_cols = [
    'STDNT_TEST_ENTRANCE_COMB', 'DISTANCE_FROM_HOME', 'HIGH_SCHL_GPA',
    'FATHER_HI_EDU_CD', 'MOTHER_HI_EDU_CD', 'SECOND_TERM_ATTEMPT_HRS', 
    'SECOND_TERM_EARNED_HRS'
]

summary_rows = []
for col in numeric_cols:
    before = df_before[col]
    after  = df[col]
    summary_rows.append({
        'col': col,
        'count_before': before.count(),
        'mean_before': before.mean(),
        'median_before': before.median(),
        'std_before': before.std(),
        'count_after': after.count(),
        'mean_after': after.mean(),
        'median_after': after.median(),
        'std_after': after.std(),
        'mean_diff': after.mean() - before.mean(),
        'median_diff': after.median() - before.median()
    })

import pandas as pd
display(pd.DataFrame(summary_rows))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of numeric columns we filled
numeric_cols = [
    'STDNT_TEST_ENTRANCE_COMB', 'DISTANCE_FROM_HOME', 'HIGH_SCHL_GPA',
    'FATHER_HI_EDU_CD', 'MOTHER_HI_EDU_CD', 'SECOND_TERM_ATTEMPT_HRS', 
    'SECOND_TERM_EARNED_HRS'
]

# Create a figure for boxplots before and after
fig, axes = plt.subplots(len(numeric_cols), 2, figsize=(15, len(numeric_cols)*4))
fig.suptitle("Comparison Before and After Filling Missing Values", fontsize=16)

for i, col in enumerate(numeric_cols):
    # Before filling 
    sns.boxplot(x=df_before[col], ax=axes[i,0], color='lightblue')
    axes[i,0].set_title(f'{col} - Before Filling')

    # After filling
    sns.boxplot(x=df[col], ax=axes[i,1], color='lightgreen')
    axes[i,1].set_title(f'{col} - After Filling')

plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.show()
```

We have some significant outliers and a few unstable columns, but we can address those later.
 
The key point for now is that the data distribution hasnt changed after filling the missing values.


# 2. Filling categorical columns with the LOCAL MEDIAN

```python
categorical_cols = df.select_dtypes(include='object').columns
df[categorical_cols].isna().sum()

```

```python
df_cat_before = df[categorical_cols].copy()
```

Filling missing COURSE NAMES using the most common value based on the students major and the semester (first or second term).

```python
df['CORE_COURSE_NAME_2_F'] = df.groupby(['FIRST_TERM','STDNT_MAJOR'])['CORE_COURSE_NAME_2_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
df['CORE_COURSE_NAME_3_F'] = df.groupby(['FIRST_TERM','STDNT_MAJOR'])['CORE_COURSE_NAME_3_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
df['CORE_COURSE_NAME_4_F'] = df.groupby(['FIRST_TERM','STDNT_MAJOR'])['CORE_COURSE_NAME_4_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
df['CORE_COURSE_NAME_1_S'] = df.groupby(['SECOND_TERM','STDNT_MAJOR'])['CORE_COURSE_NAME_1_S']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
df['CORE_COURSE_NAME_2_S'] = df.groupby(['SECOND_TERM','STDNT_MAJOR'])['CORE_COURSE_NAME_2_S']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

```python
df['CORE_COURSE_NAME_3_S'] = df.groupby(['SECOND_TERM','STDNT_MAJOR'])['CORE_COURSE_NAME_3_S']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown'))
```

Filling missing grades in COURSE GRADE using the most frequent grade for each course; if no mode exists, defaulting to 'C'.

```python
df['CORE_COURSE_GRADE_2_F'] = df.groupby('CORE_COURSE_NAME_2_F')['CORE_COURSE_GRADE_2_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
df['CORE_COURSE_GRADE_2_F'] = df.groupby('CORE_COURSE_NAME_2_F')['CORE_COURSE_GRADE_2_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
df['CORE_COURSE_GRADE_3_F'] = df.groupby('CORE_COURSE_NAME_3_F')['CORE_COURSE_GRADE_3_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
df['CORE_COURSE_GRADE_4_F'] = df.groupby('CORE_COURSE_NAME_4_F')['CORE_COURSE_GRADE_4_F']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
df['CORE_COURSE_GRADE_1_S'] = df.groupby('CORE_COURSE_NAME_1_S')['CORE_COURSE_GRADE_1_S']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
df['CORE_COURSE_GRADE_2_S'] = df.groupby('CORE_COURSE_NAME_2_S')['CORE_COURSE_GRADE_2_S']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
df['CORE_COURSE_GRADE_3_S'] = df.groupby('CORE_COURSE_NAME_3_S')['CORE_COURSE_GRADE_3_S']\
                                .transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'C'))
```

```python
results = []

for col in categorical_cols:
    before = df_cat_before[col]
    after  = df[col]
    
    results.append({
        'Column': col,
        
        # Missing counts
        'Missing Before': before.isna().sum(),
        'Missing After':  after.isna().sum(),
        
        # Mode before
        'Mode Before': before.mode()[0] if not before.mode().empty else None,
        'Mode After': after.mode()[0] if not after.mode().empty else None,
        
        # No. of unique categories
        'Unique Before': before.nunique(),
        'Unique After':  after.nunique()
    })

# Convert to DataFrame for nicer display
results_df = pd.DataFrame(results)

display(results_df)
```
