## Calculating Semester GPA

```python
sub_name = ['CORE_COURSE_NAME_1_F','CORE_COURSE_NAME_2_F','CORE_COURSE_NAME_3_F',
            'CORE_COURSE_NAME_1_S','CORE_COURSE_NAME_2_S','CORE_COURSE_NAME_3_S']

sub_grade = ['CORE_COURSE_GRADE_1_F','CORE_COURSE_GRADE_2_F','CORE_COURSE_GRADE_3_F',
            'CORE_COURSE_GRADE_1_S','CORE_COURSE_GRADE_2_S','CORE_COURSE_GRADE_3_S']
```

```python
grade_points = {
    "A": 4, "B": 3, "C": 2, "D": 1, "F": 0,
    "INCOMPL": 0, "NOT REP": 0, "Unknown": 0
}

fall_grades = sub_grade[:3]
spring_grades = sub_grade[3:]
```

```python
for g in fall_grades + spring_grades:
    df[g + "_points"] = df[g].map(grade_points)
```

```python
df["Fall_GPA"] = df[[g + "_points" for g in fall_grades]].sum(axis=1) / df[[g + "_points" for g in fall_grades]].count(axis=1)


df["Spring_GPA"] = df[[g + "_points" for g in spring_grades]].sum(axis=1) / df[[g + "_points" for g in spring_grades]].count(axis=1)
```

# Major_Category Feature Engineering

The `STDNT_MAJOR` column contains 54 unique majors, including a large number of rare majors and a dominant "Undeclared" category. 

**Solution :**  
- Created a new column `Major_Category` to group majors into broader categories based on keywords.  
- Categories include: `STEM`, `Business`, `Arts & Humanities`, `Education`, `Health`, `Social Sciences`, `Other`, and `Undeclared`.  
- This reduces complexity, and prepares the feature for analysis .


```python
df['STDNT_MAJOR'].describe()
```

```python


def categorize_major(major):
    major = str(major).lower()

    # Undeclared
    if "undeclared" in major:
        return "Undeclared"

    # STEM
    stem_keywords = ["computer", "math", "chem", "bio", "physics",
                    "geology", "engineering", "science", "applied"]
    
    if any(word in major for word in stem_keywords):
        return "STEM"

    # Business & Management
    business_keywords = ["business", "account", "finance", "marketing",
                        "management", "mis"]
    
    if any(word in major for word in business_keywords):
        return "Business"

    # Arts & Humanities
    arts_keywords = ["art", "music", "theatre", "english", "literature",
                    "french", "spanish", "arab", "history"]
    
    if any(word in major for word in arts_keywords):
        return "Arts & Humanities"
    
    # Education
    education_keywords = ["education", "teacher", "secondary", "reading"]
    
    if any(word in major for word in education_keywords):
        return "Education"
    
    # Health-related
    health_keywords = ["nursing", "exercise", "health"]
    
    if any(word in major for word in health_keywords):
        return "Health"
    
    # Social Sciences
    social_keywords = ["psychology", "crimin", "political", "sociology"]
    
    if any(word in major for word in social_keywords):
        return "Social Sciences"

    return "Other"


# Applying  categorization
df["Major_Category"] = df["STDNT_MAJOR"].apply(categorize_major)

# Show results
df["Major_Category"].value_counts()
```

### Making `FIRST_TERM` & `SECOND_TERM` more readable

```python
df['FIRST_TERM'].value_counts() , df['SECOND_TERM'].value_counts()
```

```python
def decode_term(term):
    term = str(term)
    year = term[:4]
    sem = term[4:]
    if sem == "08":
        sem_name = "Fall"
    elif sem == "02":
        sem_name = "Spring"
    else:
        sem_name = "Unknown"
    return f"{sem_name} {year}"

df['FIRST_TERM'] = df['FIRST_TERM'].apply(decode_term)
df['SECOND_TERM'] = df['SECOND_TERM'].apply(decode_term)
```
