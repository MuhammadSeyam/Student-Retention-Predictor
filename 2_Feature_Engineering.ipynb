```python
df = pd.read_csv('phase1.csv')
```

```python
# 1. Calculate the robust percentile bounds
lower_cap_percentile = df['STDNT_AGE'].quantile(0.01) # 1st percentile
upper_cap_percentile = df['STDNT_AGE'].quantile(0.99) # 99th percentile

# 2. Check the calculated bounds
print(f"1st Percentile (Lower Cap): {lower_cap_percentile}")
print(f"99th Percentile (Upper Cap): {upper_cap_percentile}")
print("-" * 30)

# 3. Apply the capping to create a new, stabilized feature
df['AGE_CAPPED'] = df['STDNT_AGE'].clip(
    lower=lower_cap_percentile,
    upper=upper_cap_percentile
)

# 4. Verify the new range
print(f"Original Max Age: {df['STDNT_AGE'].max()}")
print(f"Capped Max Age: {df['AGE_CAPPED'].max()}")
```

```python
df['AGE_CAPPED'].value_counts()
```

```python
# 1. Define the Binning Logic
# We create a dictionary to map the discrete capped ages to meaningful category names
age_bins = {
    17.0: 'Younger_Core',
    18.0: 'Main_Cohort',
    19.0: 'Older_Core'
}

# Apply the mapping to create the new categorical column 'AGE_CATEGORY'
df['AGE_CATEGORY'] = df['AGE_CAPPED'].map(age_bins)

# 2. Apply One-Hot Encoding
# The get_dummies function creates binary (0 or 1) columns for each category.
age_dummies = pd.get_dummies(
    df['AGE_CATEGORY'], 
    prefix='AGE'
)

# 3. Join the new columns back to the DataFrame
df = pd.concat([df, age_dummies], axis=1)

# Display the head of the DataFrame to show the new columns
print("New DataFrame head after Binning and One-Hot Encoding:")
print(df[['AGE_CAPPED', 'AGE_CATEGORY', 'AGE_Younger_Core', 'AGE_Main_Cohort', 'AGE_Older_Core']].head())
```

```python
df = df.drop(columns=['AGE_CATEGORY','AGE_CAPPED','STDNT_AGE'])
```

```python
# Apply One-Hot Encoding and drop the original text column
gender_dummies = pd.get_dummies(
    df['STDNT_GENDER'], 
    prefix='Gender',
    drop_first=True # Drops one category (usually 'F' since F comes before M alphabetically)
)

# Rename the remaining column for clarity
gender_dummies.rename(columns={'Gender_M': 'Is_Male'}, inplace=True)

# Concatenate the new binary column back to the main DataFrame
df = pd.concat([df, gender_dummies], axis=1)

# Drop the original text column
df.drop('STDNT_GENDER', axis=1, inplace=True)

print("New columns created and original column dropped.")
```

```python
# Define the list of categories to be grouped into 'Other'
rare_categories = ['BGD 6', 'BGD 7', 'BGD 5', 'BGD 8']

# Create a new categorical column
df['BGD_Binned'] = df['STDNT_BACKGROUND'].copy()

# Replace the rare categories with 'Other'
df['BGD_Binned'] = df['BGD_Binned'].replace(rare_categories, 'BGD_Other')

# --- 2. Apply One-Hot Encoding ---

# Create dummy variables for the binned column.
# drop_first=True prevents multicollinearity by dropping one category (e.g., BGD_1).
background_dummies = pd.get_dummies(
    df['BGD_Binned'], 
    prefix='BGD', 
    drop_first=True
)

# Concatenate the new binary columns back to the DataFrame
df = pd.concat([df, background_dummies], axis=1)

# Drop the original and intermediate columns
df.drop(['STDNT_BACKGROUND', 'BGD_Binned'], axis=1, inplace=True)

print("New binary columns for student background have been created and added to the DataFrame.")
```

```python
# Apply One-Hot Encoding (Dummy Encoding)
state_dummies = pd.get_dummies(
    df['IN_STATE_FLAG'], 
    prefix='IN_STATE',
    drop_first=True # Drops the 'IN_STATE_N' column, leaving only 'IN_STATE_Y'
)

# Rename the remaining column for clear interpretation in the model
state_dummies.rename(columns={'IN_STATE_Y': 'Is_In_State'}, inplace=True)

# Concatenate the new binary column back to the main DataFrame
df = pd.concat([df, state_dummies], axis=1)

# Drop the original text column
df.drop('IN_STATE_FLAG', axis=1, inplace=True)

print("The 'IN_STATE_FLAG' column has been successfully encoded into the 'Is_In_State' binary feature.")
```

```python
# Calculate the frequency of each major
major_counts = df['STDNT_MAJOR'].value_counts()

# Identify majors to be grouped into 'Other'
RARE_THRESHOLD = 60 
rare_majors = major_counts[major_counts < RARE_THRESHOLD].index.tolist()

# Create the binned column
df['MAJOR_Binned'] = df['STDNT_MAJOR'].copy()

# Replace all rare majors with the 'Other' label
df['MAJOR_Binned'] = df['MAJOR_Binned'].replace(rare_majors, 'Other_Major')

# --- 2. Apply One-Hot Encoding ---

# Get dummies for the binned column.
# drop_first=True prevents multicollinearity by dropping one of the categories.
major_dummies = pd.get_dummies(
    df['MAJOR_Binned'], 
    prefix='Major'
)

# Join the new binary columns back to the DataFrame
df = pd.concat([df, major_dummies], axis=1)

# Drop the original and intermediate columns
df.drop(['STDNT_MAJOR', 'MAJOR_Binned'], axis=1, inplace=True)

# You now have a set of clean, stable binary features (e.g., Major_Biology, Major_Other_Major).
print(f"Major column successfully binned (Threshold < {RARE_THRESHOLD}) and One-Hot Encoded.")
print(f"Total number of new major columns created: {len(major_dummies.columns)}")
```

```python
upper_fence = 1310
lower_fence = 670

# Apply Capping using numpy.clip
df['SCORE_CAPPED'] = df['STDNT_TEST_ENTRANCE_COMB'].clip(
    lower=lower_fence,
    upper=upper_fence)
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Reshape is needed as StandardScaler expects a 2D array
df['SCORE_SCALED'] = scaler.fit_transform(
    df[['SCORE_CAPPED']]
)
```

```python
df = df.drop(columns= ['SCORE_CAPPED','STDNT_TEST_ENTRANCE_COMB'])
```

```python
CURRENT_YEAR = 2025

# A. Convert the 'FIRST_TERM' column to string, extract the first 4 characters (the year), and convert to integer.
df['ENROLL_YEAR'] = df['FIRST_TERM'].astype(str).str[:4].astype(int)

# B. Calculate the number of years ago the student enrolled.
df['YEARS_AGO'] = CURRENT_YEAR - df['ENROLL_YEAR']

# --- 2. Scaling the New Feature ---

# Create and fit the StandardScaler
scaler = StandardScaler()

# Reshape the data and apply standardization
df['YEARS_AGO_SCALED'] = scaler.fit_transform(
    df[['YEARS_AGO']]
)

# --- 3. Clean Up ---

# Drop the original categorical column and the intermediate year column
df.drop(['FIRST_TERM', 'ENROLL_YEAR','YEARS_AGO'], axis=1, inplace=True)

# --- Verification ---
print("New feature 'YEARS_AGO_SCALED' created successfully.")
print("\nDescriptive Stats for the new feature:")
print(df['YEARS_AGO_SCALED'].describe())
```

```python
grade_map = {
    'A': 4.0, 'A-': 3.7, 'B+': 3.3, 'B': 3.0, 'B-': 2.7,
    'C+': 2.3, 'C': 2.0, 'C-': 1.7, 'D': 1.0, 'F': 0.0
}

# 2. Define Column Groups
GRADE_COLS_F = [
    'CORE_COURSE_GRADE_1_F', 'CORE_COURSE_GRADE_2_F', 
    'CORE_COURSE_GRADE_3_F', 'CORE_COURSE_GRADE_4_F'
]

GRADE_COLS_S = [
    'CORE_COURSE_GRADE_1_S', 'CORE_COURSE_GRADE_2_S', 
    'CORE_COURSE_GRADE_3_S'
]

COURSE_NAME_COLS = [
    'CORE_COURSE_NAME_1_F', 'CORE_COURSE_NAME_2_F', 'CORE_COURSE_NAME_3_F', 'CORE_COURSE_NAME_4_F',
    'CORE_COURSE_NAME_1_S', 'CORE_COURSE_NAME_2_S', 'CORE_COURSE_NAME_3_S'
]

# --- 3. Calculate First Term GPA (GPA_F) ---
df_points_F = df[GRADE_COLS_F].apply(lambda x: x.str.upper().map(grade_map))
df['GPA_F'] = df_points_F.mean(axis=1)

# --- 4. Calculate Second Term GPA (GPA_S) ---
df_points_S = df[GRADE_COLS_S].apply(lambda x: x.str.upper().map(grade_map))
df['GPA_S'] = df_points_S.mean(axis=1)

# 5. Clean Up: Drop all unnecessary columns
cols_to_drop = GRADE_COLS_F + GRADE_COLS_S + COURSE_NAME_COLS + ['SECOND_TERM']

# Drop the columns in one step
df.drop(columns=cols_to_drop, inplace=True)

print(" Two new features, 'GPA_F' and 'GPA_S', have been successfully created.")
print("The difference between these two scores (e.g., GPA_S - GPA_F) is a powerful feature for predicting student retention!")
```

```python
housing_dummies = pd.get_dummies(
    df['HOUSING_STS'], 
    prefix='HOUSING',
    drop_first=True 
)

# 2. Rename the remaining column for clarity
# The remaining column will be HOUSING_On Campus.
housing_dummies.rename(columns={'HOUSING_On Campus': 'Is_On_Campus'}, inplace=True)

# 3. Concatenate the new column and drop the original
df = pd.concat([df, housing_dummies], axis=1)
df.drop('HOUSING_STS', axis=1, inplace=True)

print("The 'HOUSING_STS' column has been successfully encoded into the 'Is_On_Campus' binary feature.")
```

```python
# Capping scores above 241.5
upper_cap = 241.5
df['DISTANCE_CAPPED'] = df['DISTANCE_FROM_HOME'].clip(upper=upper_cap)

# Transform the capped column
df['DISTANCE_LOG'] = np.log1p(df['DISTANCE_CAPPED']) # np.log1p is np.log(x + 1)

scaler = StandardScaler()
df['DISTANCE_SCALED'] = scaler.fit_transform(df[['DISTANCE_LOG']])

df = df.drop(columns=['DISTANCE_LOG','DISTANCE_CAPPED','DISTANCE_FROM_HOME'])
```

```python

# 1. Create Success Rate Features
# Handle potential division by zero by using .replace(0, 1) or a small epsilon
df['SR_F'] = df['FIRST_TERM_EARNED_HRS'] / df['FIRST_TERM_ATTEMPT_HRS'].replace(0, 1)
df['SR_S'] = df['SECOND_TERM_EARNED_HRS'] / df['SECOND_TERM_ATTEMPT_HRS'].replace(0, 1)

# 2. Create Average Load Feature
df['AVG_LOAD'] = (df['FIRST_TERM_ATTEMPT_HRS'] + df['SECOND_TERM_ATTEMPT_HRS']) / 2

# 3. Drop the redundant original columns
df.drop(columns=[
    'FIRST_TERM_ATTEMPT_HRS', 'FIRST_TERM_EARNED_HRS', 
    'SECOND_TERM_ATTEMPT_HRS', 'SECOND_TERM_EARNED_HRS'
], inplace=True)

print("Original 4 columns replaced by 3 new, less redundant features: SR_F, SR_S, AVG_LOAD.")

scale_cols = ['SR_F', 'SR_S', 'AVG_LOAD']
df[scale_cols] = scaler.fit_transform(df[scale_cols])

print("New features have been standardized and are ready for modeling.")
```

```python
# 1. Binning: Group 'A' and 'V' into 'Other'
df['DEGREE_GROUP_Binned'] = df['DEGREE_GROUP_CD'].replace(['A', 'V'], 'Other')

# 2. Apply Dummy Encoding
# drop_first=True will drop the 'Other' column, leaving the 'B' column.
degree_dummies = pd.get_dummies(
    df['DEGREE_GROUP_Binned'], 
    prefix='DEGREE',
    drop_first=True 
)

# 3. Rename and Concatenate
# The resulting column is 'DEGREE_B' (since 'B' comes after 'Other' alphabetically in this case)
df.rename(columns={'DEGREE_B': 'Is_Degree_Group_B'}, inplace=True)
df = pd.concat([df, degree_dummies], axis=1)

# 4. Drop original columns
df.drop(['DEGREE_GROUP_CD', 'DEGREE_GROUP_Binned'], axis=1, inplace=True)

print("The DEGREE_GROUP_CD column is now the stable binary feature 'Is_Degree_Group_B'.")
```
